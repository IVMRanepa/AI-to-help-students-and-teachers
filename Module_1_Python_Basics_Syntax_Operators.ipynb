# ========== –†–ê–ó–î–ï–õ 1: –ë–ê–ó–û–í–´–ï –¢–ò–ü–´ –î–ê–ù–ù–´–• –û–ü–ï–†–ê–¢–û–†–´ ==========

# ========== –†–ê–ó–î–ï–õ 3: [–ì–ï–ù–ï–†–ê–¢–û–†–´, –î–ï–ö–û–†–ê–¢–û–†–´](https://docs.google.com/presentation/d/1fVHnDg4ExFJAQfyQ7fBuLn_XbvP50Hk56hb5nnXo1yY/edit?usp=sharing), *args/**kwargs ==========

```python
print("\n\n" + "="*60)
print("–†–ê–ó–î–ï–õ 3: –ì–ï–ù–ï–†–ê–¢–û–†–´ –ò –î–ï–ö–û–†–ê–¢–û–†–´")
print("="*60)

# 1. List Comprehensions
print("\n1. LIST COMPREHENSIONS")
print("-"*60)

# –ü—Ä–æ—Å—Ç–∞—è list comprehension
squares = [x**2 for x in range(1, 6)]
print(f"–ö–≤–∞–¥—Ä–∞—Ç—ã: {squares}")

# –° —É—Å–ª–æ–≤–∏–µ–º
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(f"–ö–≤–∞–¥—Ä–∞—Ç—ã —á—ë—Ç–Ω—ã—Ö: {even_squares}")

# —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ —Ü–∏–∫–ª–∞–º–∏
matrix = [[i+j for j in range(3)] for i in range(3)]
print(f"–ú–∞—Ç—Ä–∏—Ü–∞:\n{np.array(matrix)}")

# 2. –ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã
print("\n2. –ì–ï–ù–ï–†–ê–¢–û–†–´ (yield)")
print("-"*60)

def fibonacci_gen(n):
    """Generator for Fibonacci numbers"""
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print("–ü–µ—Ä–≤—ã–µ 10 —á–∏—Å–µ–ª Fibonacci:")
fib_list = list(fibonacci_gen(10))
print(fib_list)

# 3. *args –∏ **kwargs
print("\n3. *args –∏ **kwargs")
print("-"*60)

def process_data(*args, **kwargs):
    """Process variable arguments and keyword arguments"""
    print(f"–ü–æ–∑–∏—Ü–∏–æ–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (*args): {args}")
    print(f"–ö–ª—é—á–µ–≤—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (**kwargs): {kwargs}")

process_data(1, 2, 3, name="Alice", age=30, city="NYC")

# 4. –î–µ–∫–æ—Ä–∞—Ç–æ—Ä—ã
print("\n4. –î–ï–ö–û–†–ê–¢–û–†–´")
print("-"*60)

def timer_decorator(func):
    """Decorator to measure execution time"""
    def wrapper(*args, **kwargs):
        import time
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è: {end - start:.4f} —Å–µ–∫.")
        return result
    return wrapper

@timer_decorator
def slow_function():
    """A slow function for demonstration"""
    import time
    time.sleep(0.1)
    return "Done!"

result = slow_function()
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {result}")
```

```python
print("\n2. –ö–û–ù–¢–ï–ù–ï–†–ù–´–ï –¢–ò–ü–´: list, tuple, dict, set")
print("-"*60)

# –°–µ–Ω–µ—Ä–∏—Ä—É—é–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Ä–∞–±–æ—Ç—ã
dates_list = [datetime(2025, 1, i+1) for i in range(5)]
prices = [random.uniform(10, 100) for _ in range(5)]
names = ["Alice", "Bob", "Charlie", "David", "Eve"]

# –°–ø–∏—Å–æ–∫ (list) - –º–∏—Ä–∂—É—ë–º—ã–µ –¥–∞–Ω–Ω—ã–µ
list_demo = names.copy()
print(f"list: {list_demo}")
print(f"  –ù–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç: {list_demo[0]}")
print(f"  –û—Å—Ç–∞–Ω–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç: {list_demo[-1]}")
print(f"  –û—Å—Ç–∞–Ω–Ω–∏–µ 3 —ç–ª–µ–º–µ–Ω—Ç–∞: {list_demo[-3:]}")

# –ö–æ—Ä—Ç–µ–∂ (tuple) - –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–µ –¥–∞–Ω–Ω—ã–µ
tuple_demo = tuple(prices[:3])
print(f"\ntuple: {tuple_demo}")
print(f"  –î–ª–∏–Ω–∞: {len(tuple_demo)}")

# –°–ª–æ–≤–∞—Ä—å (dict) - –¥–∞–Ω–Ω—ã–µ –∫ –∫–ª—é—á–∞–º
dict_demo = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "salary": prices[0]
}
print(f"\ndict: {dict_demo}")
print(f"  –ö–ª—é—á–∏: {list(dict_demo.keys())}")
print(f"  –ó–Ω–∞—á–µ–Ω–∏—è: {list(dict_demo.values())}")

# –ú–Ω–æ–∂–µ—Å—Ç–≤–æ (set) - —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
set_demo = set(["Python", "Java", "Python", "C++", "Java"])
print(f"\nset: {set_demo}")
print(f"  –ú–æ—â–Ω–æ—Å—Ç—å: {len(set_demo)}")
```

```python
print("\n3. –û–ü–ï–†–ê–¢–û–†–´")
print("-"*60)

# –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã
a, b = 15, 4
print(f"\n–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã (a={a}, b={b}):")
print(f"  a + b = {a + b}  # —Å–ª–æ–∂–µ–Ω–∏–µ")
print(f"  a - b = {a - b}  # –≤—ã—á–∏—Ç–∞–Ω–∏–µ")
print(f"  a * b = {a * b}  # —É–º–Ω–æ–∂–µ–Ω–∏–µ")
print(f"  a / b = {a / b}  # –¥–µ–ª–µ–Ω–∏–µ")
print(f"  a // b = {a // b}  # —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ")
print(f"  a % b = {a % b}  # –æ—Å—Ç–∞—Ç–æ–∫")
print(f"  a ** b = {a ** b}  # –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å")

# –û–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
print(f"\n–û–ø–µ—Ä–∞—Ç–æ—Ä—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è:")
print(f"  {a} > {b} => {a > b}")
print(f"  {a} < {b} => {a < b}")
print(f"  {a} == {b} => {a == b}")
print(f"  {a} != {b} => {a != b}")
print(f"  {a} >= {b} => {a >= b}")
print(f"  {a} <= {b} => {a <= b}")

# –õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã
cond1, cond2 = True, False
print(f"\n–õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã (cond1={cond1}, cond2={cond2}):")
print(f"  cond1 and cond2 => {cond1 and cond2}")
print(f"  cond1 or cond2 => {cond1 or cond2}")
print(f"  not cond1 => {not cond1}")
print(f"  not cond2 => {not cond2}")
```

# ========== –†–ê–ó–î–ï–õ 2: –§–£–ù–ö–¶–ò–ò –ò –£–ü–†–ê–í–õ–ï–ù–ò–ï –ü–û–¢–û–ö–û–ú ==========

```python
print("\n\n" + "="*60)
print("–†–ê–ó–î–ï–õ 2: –§–£–ù–ö–¶–ò–ò –ò –£–ü–†–ê–í–õ–ï–ù–ò–ï –ü–û–¢–û–ö–û–ú")
print("="*60)

# 1. –û—Å–Ω–æ–≤–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏
print("\n1. –û—Å–Ω–æ–≤–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏")
print("-"*60)

# –ü—Ä–æ—Å—Ç–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def calculate_average(numbers):
    """
    –ö–∞–ª—å–∫—É–ª–∏—Ä—É–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –¥–∞–Ω–Ω—ã—Ö.
    """
    return sum(numbers) / len(numbers)

data = [10, 20, 30, 40, 50]
avg = calculate_average(data)
print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {avg}")

# –§—É–Ω–∫—Ü–∏—è —Å –≤—ã–∑–≤–∞–Ω–Ω—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º
def greet(name, age=25):
    """
    –ü—Ä–∏–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ø–æ –∏–º–µ–Ω–∏ –∏ –≤–æ–∑—Ä–∞—Å—Ç—É.
    """
    return f"Hello, {name}! You are {age} years old."

print(f"\n{greet('Alice')}")
print(f"{greet('Bob', 30)}")

# –§—É–Ω–∫—Ü–∏—è —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –≤—ã—Ö–æ–¥–Ω—ã–º–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
def calculate_stats(numbers):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∏–Ω–∏–º—É–º, –º–∞–∫—Å–∏–º—É–º, —Å—Ä–µ–¥–Ω–µ–µ.
    """
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

min_val, max_val, avg_val = calculate_stats([5, 15, 25, 35, 45])
print(f"\n–ú–∏–Ω: {min_val}, –ú–∞–∫—Å: {max_val}, –°—Ä–µ–¥: {avg_val:.2f}")
```

# 2. –£–ü–†–ê–í–õ–ï–ù–ò–ï –ü–û–¢–û–ö–û–ú: if/else/elif, for, while

```python
print("\n2. –£–ü–†–ê–í–õ–ï–ù–ò–ï –ü–û–¢–û–ö–û–ú")
print("-"*60)

# if/else/elif –æ–±—ã—á–Ω—ã–µ –ø—Ä–∏–º–µ—Ä—ã
print("\nif/else/elif:")
for score in [45, 55, 75, 85, 95]:
    if score < 50:
        grade = "F"
    elif score < 60:
        grade = "D"
    elif score < 70:
        grade = "C"
    elif score < 80:
        grade = "B"
    else:
        grade = "A"
    print(f"  Score: {score} => Grade: {grade}")

# for —Ü–∏–∫–ª
print("\nfor loop:")
print("  –ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 10:")
for i in range(1, 11):
    print(f"    {i}", end=" ")
print()

# for —Å —ç–Ω—É–º–µ—Ä–∞—Ü–∏–µ–π
print("\n  –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ø–∏—Å–∫–∞:")
for idx, name in enumerate(["Alice", "Bob", "Charlie"], 1):
    print(f"    {idx}. {name}")

# while —Ü–∏–∫–ª
print("\nwhile loop:")
count = 0
while count < 5:
    print(f"  Count: {count}")
    count += 1

# –¶–∏–∫–ª —Å break/continue
print("\nfor loop —Å break/continue:")
for i in range(10):
    if i == 3:
        continue  # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º 3
    if i == 7:
        break  # –û—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø—Ä–∏ 7
    print(f"  {i}", end=" ")
print()
```
