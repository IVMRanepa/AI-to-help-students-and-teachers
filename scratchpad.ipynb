# 1. ГЕНЕРАТОРЫ - ЗАДАЧА 1 (Генератор Фибоначчи)

```
# 1. ГЕНЕРАТОРЫ - ЗАДАЧА 1 (Генератор Фибоначчи)

print('\n' + '='*60)
print('ГЕНЕРАТОРЫ - ЗАДАЧА 1')
print('='*60)
print('Генератор последовательности Фибоначчи\n')

# Определяем генератор для последовательности Фибоначчи
def fibonacci_generator(n):
    """
    Генератор, который генерирует первые n чисел Последовательности Фибоначчи
    """
    # Коварианты запуска
    a, b = 0, 1
    count = 0
    # Цикл до нормальности count величины n
    while count < n:
        # yield прерывает выполнение и возвращает значение
        yield a
        a, b = b, a + b  # Обновляем пару (а, b)
        count += 1

# Пример использования: генерируем первые 10 чисел Фибоначчи
print('Пример 1: Первые 10 чисел Фибоначчи:')
fib_sequence = list(fibonacci_generator(10))
print(fib_sequence)
print(f'Длина последовательности: {len(fib_sequence)}')

# Пример 2: Обработка генератора в цикле
print('\nПример 2: Обработка в цикле:')
for i, fib_num in enumerate(fibonacci_generator(8)):
    print(f'Приток {i+1}: {fib_num}')

# ВЫВОДЫ для МОДУЛЯ 1, ЗАДАЧА 1:
print('\nВЫВОДЫ: Генераторы в Python полезны для:')
print('  - Энергоэффективного использования памяти для больших последовательностей')
print('  - Обработки бесконечных потоков данных')
print('  - Наления ты "ленивой" вычисляемости (lazy evaluation)')
```

# 1. ГЕНЕРАТОРЫ - ЗАДАЧА 2 (Пакеты данных для ML)

```
# 1. ГЕНЕРАТОРЫ - ЗАДАЧА 2 (Пакеты данных для ML)

print('\n' + '='*60)
print('ГЕНЕРАТОРЫ - ЗАДАЧА 2')
print('='*60)
print('Генератор батчей данных для машинного обучения\n')

def data_batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

test_data = list(range(1, 26))
print(f'Общие данные: {test_data}')
print(f'Тотально элементов: {len(test_data)}\n')

print('Пример 1: Обработка батчами по 5:')
for batch_num, batch in enumerate(data_batch_generator(test_data, 5), 1):
    print(f'  Батч {batch_num}: {batch}')

print('\nВЫВОДЫ: Генераторы важны в ML для экономии памяти')
```

# 2. ДЕКОРАТОРЫ - ЗАДАЧА 1 (Мерило времени)

```
# 2. ДЕКОРАТОРЫ - ЗАДАЧА 1 (Мерило времени)

import time
from functools import wraps

print('\n' + '='*60)
print('ДЕКОРАТОРЫ - ЗАДАЧА 1')
print('='*60)
print('Декоратор для измерения времени выполнения\n')

# Определяем декоратор для меря времени
def timer_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        старт = time.time()
        result = func(*args, **kwargs)
        конец = time.time()
        time_taken = конец - старт
        print(f'  \u0412ремя выполнения {func.__name__}: {time_taken:.4f} секунд')
        return result
    return wrapper

# Применяем декоратор
@timer_decorator
def calculate_sum(n):
    сумма = sum(range(n))
    return сумма

@timer_decorator
def slow_function():
    time.sleep(0.1)
    return 'На нарочно медленная функция'

print('Пример 1: Мера времени для calculate_sum:')
result = calculate_sum(1000000)
print(f'  Результат: {result}\n')

print('Пример 2: Мера времени для slow_function:')
result2 = slow_function()
print(f'  Результат: {result2}')

print('\nВЫВОДЫ: Декораторы позволяют профилировать код без изменения исходных функций')
```

# 2. ДЕКОРАТОРЫ - ЗАДАЧА 2 (Валидация аргументов)

```
# 2. ДЕКОРАТОРЫ - ЗАДАЧА 2 (Валидация аргументов)

print('\n' + '='*60)
print('ДЕКОРАТОРЫ - ЗАДАЧА 2')
print('='*60)
print('Декоратор для валидации типов аргументов\n')

# Определяем декоратор валидации
def validate_types(**types_dict):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            # Проверяем типы kwargs
            for key, value in kwargs.items():
                if key in types_dict and not isinstance(value, types_dict[key]):
                    raise TypeError(f'{key} должен быть {types_dict[key].__name__}, получен {type(value).__name__}')
            return func(*args, **kwargs)
        return wrapper
    return decorator

# Применяем декоратор
@validate_types(age=int, name=str)
def create_user(name, age):
    return f'Пользователь {name}, возраст {age}'

print('Пример 1: Корректные аргументы:')
try:
    result = create_user(name='Иван', age=25)
    print(f'  Успех: {result}')
except TypeError as e:
    print(f'  Ошибка: {e}')

print('\nПример 2: Некорректные типы:')
try:
    result = create_user(name='Мария', age='30')
    print(f'  Успех: {result}')
except TypeError as e:
    print(f'  Ошибка: {e}')

print('\nВЫВОДЫ: Декораторы типизации улучшают надежность кода')
```

# 3. НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 1 (*args)

```
# 3. НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 1 (*args)

print('\n' + '='*60)
print('НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 1')
print('='*60)
print('*args: Принимает любое количество позиционных аргументов\n')

# Определяем функцию объединения списков
def merge_lists(*lists):
    """Объединяет любое количество списков"""
    result = []
    for lst in lists:
        result.extend(lst)
    return result

print('Пример 1: Объединение 3 списков:')
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = [7, 8, 9]
result = merge_lists(list1, list2, list3)
print(f'  {list1} + {list2} + {list3} = {result}')

print('\nПример 2: Объединение 5 списков:')
result2 = merge_lists([1], [2, 3], [4], [5, 6], [7, 8, 9])
print(f'  Объединенный результат: {result2}')

print('\nВЫВОДЫ: *args очень выступает гибкостью аргументов')
```

# 3. НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 2 (**kwargs)

```
# 3. НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 2 (*kwargs*)

print('\n' + '='*60)
print('НЕОГРАНИЧЕННЫЕ АРГУМЕНТЫ - ЗАДАЧА 2')
print('='*60)
print('**kwargs: Принимает именованные аргументы в виде словаря\n')

# Определяем функцию конфигурации
def configure_server(**config):
    """Конфигурирует сервер с любыми параметрами"""
    defaults = {'host': 'localhost', 'port': 8000, 'debug': False}
    defaults.update(config)
    return defaults

print('Пример 1: Конфиг со стандартными параметрами:')
config1 = configure_server()
for key, value in config1.items():
    print(f'  {key}: {value}')

print('\nПример 2: Конфиг с кустомными параметрами:')
config2 = configure_server(host='0.0.0.0', port=5000, debug=True, workers=4)
for key, value in config2.items():
    print(f'  {key}: {value}')

print('\nВЫВОДЫ: **kwargs текучи для содая функций с неограниченным числом конфиг опций')
```
